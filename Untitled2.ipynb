```python
url='https://raw.githubusercontent.com/Sinchiguano/ai_googleColab/main/datasetUleamMl/IRIS.csv'
import pandas as pd
dataset=pd.read_csv(url)
```

```python
dataset.head(10)
```

```python
dataset['species'].unique()
```

```python
dataset.shape
```

```python
dataset['species'].value_counts()
```

```python
tmp=list()
for etiqueta in dataset['species']:
  if etiqueta =='Iris-setosa':
    tmp.append(1)
  elif etiqueta == 'Iris-versicolor':
    tmp.append(2)
  else:
    tmp.append(3)

dataset['etiqueta']=tmp
```

```python
dataset.head(10)
```

```python
dataset.tail(5)
```

```python
dataset['etiqueta'].unique()
```

```python
dataset.groupby('species').size()
```

```python
#array(['Iris-setosa','Iris-versicolor','Iris-virginica'],dtype=object)
maskSetosa=dataset['species']=='Iris-setosa'
maskVersicolor=dataset['species']=='Iris-versicolor'
maskVirginica=dataset['species']=='Iris-virginica'

setosa=dataset[maskSetosa]
versicolor=dataset[maskVersicolor]
virginica=dataset[maskVirginica]
```

```python
setosa.head(3)
```

```python
setosa.tail(3)
```

```python
print(setosa.shape)
print(versicolor.shape)
print(virginica.shape)
```

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
fig.set_size_inches(14,6)

ax.scatter(setosa['sepal_length'],setosa['sepal_width'],label='setosa')
ax.scatter(versicolor['sepal_length'],versicolor['sepal_width'],label='versicolor')
ax.scatter(virginica['sepal_length'],virginica['sepal_width'],label='virginica')

ax.legend()
```

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
fig.set_size_inches(14,6)

ax.scatter(setosa['petal_length'],setosa['petal_width'],label='setosa')
ax.scatter(versicolor['petal_length'],versicolor['petal_width'],label='versicolor')
ax.scatter(virginica['petal_length'],virginica['petal_width'],label='virginica')

ax.legend()
```
